## Statistical Test

### Fine whether there is any significant different in consistency rates among strikers from various nationality

```python
# Normality test

stat, p_value = shapiro(data['Consistency'])
print('P value: ', p_value)
```

```python
# Filtering data

brazilian = data.query('Nationality == "Brazil"')['Consistency']
englander = data.query('Nationality == "England"')['Consistency']
frence = data.query('Nationality == "France"')['Consistency']
german = data.query('Nationality == "Germany"')['Consistency']
spaniard = data.query('Nationality == "Spain"')['Consistency']
```

```python
# levene test for statistics

alpha = 0.5
stats, p_value = levene(brazilian, englander, frence, german, spaniard)
print('P value: ', p_value)

if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
```

```python
# one way ANOVA

test_stat, p_value = stats.f_oneway(brazilian, englander, frence, german, spaniard)
print('P value: ', p_value)

if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
```

### Check if there is any significant correlation between strikers Hold-up play and consistency rate

```python
# Normality test

stats, p_value = shapiro(data['Hold-up Play'])
print('P value: ', p_value)

if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
```

```python
# Linearity test

plt.figure(figsize= (10, 6))
sns.regplot(x = 'Hold-up Play', y= 'Consistency', data=data)
plt.title('Correlation between Hold-up Play and Consistency')
plt.xlabel('Hold-up Play')
plt.ylabel('Consistency')
plt.show()
```

```python
holdup = data['Hold-up Play']
consist = data['Consistency']

corr, p_value = pearsonr(holdup, consist)
print('Correlation coefficient: ', corr)
print('P value: ', p_value)

if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
```

## Featuring Engineering

### Creating new feature - total contribution score

```python
data['total_cs'] = (data['Goals Scored'] + data['Assists'] + data['Shots on Target'] + data['Dribbling Success'] + data['Aerial Duels Won'] + data['Defensive Contribution'] + data['Big Game Performance'] + data['Consistency'])
data.head()
```

### Encode the footedness and marital status by LabelEncoder

```python
encoder = LabelEncoder()
data['Footedness'] = encoder.fit_transform(data['Footedness'])
data['Marital Status'] = encoder.fit_transform(data['Marital Status'])
data.head()
```

### Create dummies for Nationality and add with the data

```python
dummies = pd.get_dummies(data['Nationality'])
processed_df = pd.concat([data, dummies], axis=1)
processed_df = processed_df.drop('Nationality', axis= 1)
processed_df.head()
```

## Clustering Analysis

### Perform KMeans clustering

```python
# Select featuring
x = processed_df.drop('Striker_ID', axis= 1)

# Calculating WCSS score
wcss = []

for i in range(1, 15):
    kmeans = KMeans(n_clusters= i, init='k-means++' )
    kmeans.fit(x)
    wcss_score = kmeans.inertia_
    wcss.append(wcss_score)
```

```python
# Plotting the elbow chart

plt.figure(figsize= (12, 6))
plt.plot(range(1, 15), wcss, marker = 'o')
plt.title('Elbow method')
plt.xlabel('number of clusters (n)')
plt.ylabel('wcss')
plt.show()
```

```python
# Building KMeans with k

final_km = KMeans(n_clusters=2)
final_km.fit(x)

# Generating labels
labels = final_km.labels_
labels
```

```python
# Adding labels

processed_df['Clusters'] = labels
processed_df.head()
```

```python
# Checking clusters
round(processed_df.groupby('Clusters')['total_cs'].mean(), 2)
```

```python
# Assigning meaningfull name
mapping = {1 : 'Best Strikers', 0 : 'Regular Strikers'}
processed_df['Strikers types'] = processed_df['Clusters'].map(mapping)
processed_df.head()
```

## Machine Lerning Model

### Selecting features

```python
x = processed_df.drop(['Strikers types'], axis= 1)
y = processed_df['Strikers types']
```

### Scaling features

```python
scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
scaled_x
```

### Train test split

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size = 0.2, random_state = 42)
```

## Predictive Classification Analytics

### Build a logistic regression machine leaning model to predict strikers type

```python
# Model training
lgr_model = LogisticRegression()
lgr_model.fit(x_train, y_train)

# Prediction
y_lgr_pred = lgr_model.predict(x_test)

# Evaluation
accuracy_lgr = accuracy_score(y_test, y_lgr_pred)
print (accuracy_lgr* 100, '%')
```

```python
# Creating confusion matrix
conf_matrix_lgr = confusion_matrix(y_test, y_lgr_pred)

# Plotting confusion matrix
plt.figure(figsize = (12, 6))
sns.heatmap(conf_matrix_lgr, annot = True, fmt = "d", cmap = "Blues")
plt.title('Confusion Matrix for LGR model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
